```python
# Задание глобальных переменных 

# путь до файлов с данными
DS_FILES_PATH = "/user/bigdata-course/PAMAP2 Physical Activity Monitoring/PAMAP2_Dataset/Protocol/*.dat" 

# разделитель - пробел
DELIMETER = " "



# количество столбцов из описания набора данных
COLUMNS_COUNT = 54

# названия столбцов из описания набора данных
ds_columns = ["timestamp", "activityID", "heart_rate",
"IMU_hand_temperature",
"IMU_hand_acc_16g_x", "IMU_hand_acc_16g_y", "IMU_hand_acc_16g_z", "IMU_hand_acc_6g_x", "IMU_hand_acc_6g_y", "IMU_hand_acc_6g_z",
"IMU_hand_gyr_x", "IMU_hand_gyr_y", "IMU_hand_gyr_z",
"IMU_hand_mag_x", "IMU_hand_mag_y", "IMU_hand_mag_z",
"IMU_hand_ori_0", "IMU_hand_ori_1", "IMU_hand_ori_2", "IMU_hand_ori_3", "IMU_chest_temperature",
"IMU_chest_acc_16g_x", "IMU_chest_acc_16g_y", "IMU_chest_acc_16g_z", "IMU_chest_acc_6g_x", "IMU_chest_acc_6g_y", "IMU_chest_acc_6g_z", "IMU_chest_gyr_x", 
              "IMU_chest_gyr_y", "IMU_chest_gyr_z",
"IMU_chest_mag_x", "IMU_chest_mag_y", "IMU_chest_mag_z",
"IMU_chest_ori_0", "IMU_chest_ori_1", "IMU_chest_ori_2", "IMU_chest_ori_3", "IMU_ankle_temperature",
"IMU_ankle_acc_16g_x", "IMU_ankle_acc_16g_y", "IMU_ankle_acc_16g_z", "IMU_ankle_acc_6g_x", "IMU_ankle_acc_6g_y", "IMU_ankle_acc_6g_z", "IMU_ankle_gyr_x", 
              "IMU_ankle_gyr_y", "IMU_ankle_gyr_z",
"IMU_ankle_mag_x", "IMU_ankle_mag_y", "IMU_ankle_mag_z",
"IMU_ankle_ori_0", "IMU_ankle_ori_1", "IMU_ankle_ori_2", "IMU_ankle_ori_3"]

# проверяем, что не ошиблись в количестве при вводе 
assert len(ds_columns) == COLUMNS_COUNT
```

```python
# Чтение исходного набора данных в DataFrame

# читаем наборы данных из файлов 
rdd = sc.textFile(DS_FILES_PATH) 

numbers_rdd = rdd.map(lambda x : list(map(float, x.split(DELIMETER)))) 

df = numbers_rdd.toDF(ds_columns)

df.show(1, vertical=True)
```

```python
#Отображение статистики по избранным столбцам
df.select("activityID", "heart_rate", "IMU_hand_temperature",
"IMU_chest_temperature", "IMU_ankle_temperature").describe().show()
```

```python
df_no_NaNs = df.dropna() 
df_no_NaNs.select("activityID", "heart_rate",
"IMU_hand_temperature", "IMU_chest_temperature", "IMU_ankle_temperature").describe().show()
```

```python
df_prepared = df_no_NaNs.drop("timestamp",
"IMU_hand_ori_0", "IMU_hand_ori_1", "IMU_hand_ori_2", "IMU_hand_ori_3", "IMU_chest_ori_0", "IMU_chest_ori_1", "IMU_chest_ori_2", "IMU_chest_ori_3", 
                              "IMU_ankle_ori_0", 
                              "IMU_ankle_ori_1", "IMU_ankle_ori_2", "IMU_ankle_ori_3")
df_prepared.show(1, vertical=True)
```

```python
df_train, df_test = df_prepared.randomSplit([0.8, 0.2])
assert df_train.count() + df_test.count() == df_prepared.count() 
print(df_train.count())
print(df_test.count())
```

```python
#Разделение записей на пары меток и списка признаков 

from pyspark.mllib.regression import LabeledPoint

train_rdd = df_train.rdd
test_rdd = df_test.rdd

labeled_train_rdd = train_rdd.map(lambda row : LabeledPoint(row[0], row[1:])) 
labeled_test_rdd = test_rdd.map(lambda row : LabeledPoint(row[0], row[1:])) 
labeled_train_rdd.first()
```

```python
#Создание и обучение дерева решений
from time import time

from pyspark.mllib.tree import DecisionTree, DecisionTreeModel 
from pyspark.mllib.util import MLUtils
start_time = time()
model = DecisionTree.trainClassifier( labeled_train_rdd,
numClasses=25, categoricalFeaturesInfo={}, maxDepth=30,
maxBins=32)
end_time = time()
duration = end_time - start_time
print("Время обучения: %.3f секунд" % duration)
```

```python
#Оценка точности и структура модели


start_time = time()
predictions = model.predict(labeled_test_rdd.map(lambda x: x.features))
end_time = time() 
labels_and_predictions = labeled_test_rdd\
.map(lambda lp: lp.label)\
.zip(predictions)
tp = labels_and_predictions\
.filter(lambda lp: lp[0] == lp[1])\
.count()
fn = labels_and_predictions\
.filter(lambda lp: lp[0] != 0 and lp[1] ==0)\
.count()
fp = labels_and_predictions\
.filter(lambda lp: lp[0] == 0 and lp[1] != 0)\
.count()

accuracy_percent = round(tp * 100 / float(labeled_test_rdd.count()),3)
print("Доля верных предсказаний(accuracy):", str(accuracy_percent) + "%") 

precision = round((tp*100/(tp+fp)),3)
print("Precision:", str(precision) + "%")

recall = round((tp*100/(tp+fn)),3)
print("Recall: " , str(recall) + "%")

Fmera = round(2*precision*recall/(precision+recall),3)
print("F-мера:" , str(Fmera)+"%")

duration = end_time - start_time

print("Время прогнозирования: %.3f секунд" % duration)

print("Структура обученной модели:")

print(model.toDebugString())
```
